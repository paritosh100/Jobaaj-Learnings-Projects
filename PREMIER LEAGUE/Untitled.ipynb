```python
# Get a list of unique club names from both 'HomeTeam' and 'AwayTeam' columns
home_club_names = copy_df['HomeTeam'].unique()
away_club_names = copy_df['AwayTeam'].unique()
club_names = sorted(set(home_club_names) | set(away_club_names))

# Create a dictionary to store the data for each club
club_data = {}

# Iterate over each club and extract their data
for club_name in club_names:
    home_data = copy_df[copy_df['HomeTeam'] == club_name]
    away_data = copy_df[copy_df['AwayTeam'] == club_name]
    club_data[club_name] = pd.concat([home_data, away_data])

# Save the data for each club in separate CSV files
for club_name, data in club_data.items():
    file_name = f'{club_name}_data.csv'
    data.to_csv(file_name, index=False)
```
