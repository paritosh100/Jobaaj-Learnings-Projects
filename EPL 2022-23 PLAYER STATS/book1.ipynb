```python
passing_85 = og_copy[og_copy['FKAccuracy']>85]
passing_85.sort_values(by="FKAccuracy",ascending=False)
```

```python
fig = plt.figure(figsize=(20,10))
sns.barplot(data=passing_85,x = "Name",y = "PassingTotal")
plt.xticks(rotation = 90,fontsize =20)
plt.xlabel("NAMES",fontsize=20)
plt.ylabel('PASSING ACCURACY',fontsize=20)
for i, value in enumerate(passing_85['PassingTotal']):
    plt.annotate(str(value),(i,value),ha = 'center',va = 'bottom',fontsize = 20)
plt.show()
```

```python
fig = plt.figure(figsize=(20,10))
passing_85.sort_values(by = "FKAccuracy",ascending=False)
sns.barplot(data=passing_85,x = "Name",y = "FKAccuracy")
plt.xticks(rotation = 90,fontsize =20)
plt.xlabel("NAMES",fontsize=20)
plt.grid(True)
plt.ylabel('FREE KICK ACCURACY',fontsize=20)
for i, value in enumerate(passing_85['FKAccuracy']):
    plt.annotate(str(value),(i,value),ha = 'center',va = 'bottom',fontsize = 20)
plt.show()
```
