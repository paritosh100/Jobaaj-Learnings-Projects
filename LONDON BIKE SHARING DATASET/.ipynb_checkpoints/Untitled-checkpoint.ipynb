```python
#since the weather code and few columns have numerical data which can cause confusion at times, we shall replace them with real values
og_copy = df.copy()
```

```python
og_copy['weather_code'] = og_copy['weather_code']\
                            .replace({1.0:'Clear',
                                      2.0:'Few Clouds',
                                      3.0:'Broken Clouds',
                                     4.0:'Cloudy',
                                     7.0:'Light Rain',
                                     10.0:'Thunderstorm + Rain',
                                     26.0:'Snowfall',
                                     94.0:'Freezing Fog'}) 
```

```python
og_copy
```

```python
og_copy['season'] = og_copy['season']\
                            .replace({0.0:'Spring',
                                      1.0:'Summer',
                                      2.0:'Fall',
                                     3.0:'Winter'
                                    }) 
```

```python
winter_data = og_copy[og_copy['season'] == 'Winter']
summer_data = og_copy[og_copy['season'] == 'Summer']
spring_data = og_copy[og_copy['season'] == 'Spring']
fall_data = og_copy[og_copy['season'] == 'Fall']
winter_data
```

# How does weather affect bike shares?

```python
og_weather_plot = pd.DataFrame(og_copy.groupby('weather_code')['cnt'].agg('sum'))
```

```python
og_weather_plot = og_weather_plot.sort_values('cnt',ascending=False)
og_weather_plot
```

```python
fig = plt.figure(figsize=(15,8))
# Plot the values
plt.bar(og_weather_plot.index,og_weather_plot['cnt'])
plt.xlabel('WEATHER CODE',fontsize = 20)
plt.ylabel('NUMBER OF BIKE SHARES (*10Lakhs)',fontsize = 20)
plt.title('NUMBER OF BIKE SHARES PER WEATHER CODE',fontsize = 20)
plt.xticks(fontsize = 13)  
plt.yticks(fontsize = 20)  # Rotate x-axis labels if needed
for i , value in enumerate(og_weather_plot['cnt']):
    plt.annotate(str(value),(i,value),ha = 'center',va='bottom',fontsize = 15)
plt.show()
```

## When the weather is clear there is higher probability of users opting for bike sharing
